```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/bank-additional-full.csv',sep=";")
df.head()
```

```python
df.describe()
```

```python
df.info() #we can see there are no null values
```

```python
df['y'] = [1 if val=="yes" else 0 for val in df['y']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Create histogram of age
sns.kdeplot(data=df, x='age', hue="y")
plt.title("Age Densities (1 if yes, 0 if no)",fontsize=16)
sns.despine()
```

```python
df_g = df.pivot_table(index='job',values='y',columns="education",aggfunc='mean').reset_index()
df_g = df_g.melt(id_vars='job',value_name='value',var_name='education')
df_g
```

```python
#create bar plot of job and count of "y"==1

ax,fig = plt.subplots(figsize=(14,4))
sns.barplot(df_g,x='job',y='value',hue='education')
plt.xticks(fontsize=8)
plt.xlabel("Job Type")
plt.ylabel("Count of 'Yes'")
plt.title("'Yes' Count vs. Job Type and Education")
sns.despine()
```

```python
df_g_day = df.pivot_table(index="month",values="y",columns="day_of_week",aggfunc='mean').reset_index()
df_g_day = df_g_day.melt(id_vars = "month",var_name="day_of_week",value_name = "values")
```

```python
ax,fig = plt.subplots(figsize=(15,6))
sns.barplot(df_g_day,x="month",y="values",hue="day_of_week")
plt.xticks(fontsize=8)
plt.xlabel("Month")
plt.ylabel("Count of 'Yes'")
plt.title("'Yes' Count vs. Month and Day of Week")
sns.despine()
```
